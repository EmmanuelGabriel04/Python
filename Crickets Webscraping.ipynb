```python
#import the dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import datetime
import time
```

```python
url = 'https://www.bcci.tv/stats?platform=international&type=women'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
header = table.find_all('tr')[0]
```

```python
header2 = header.find_all('span')
```

```python
print(header2)
```

```python
header2_tittles = [title.text for title in header2]
print(header2_tittles)
```

```python
df = pd.DataFrame(columns = header2_tittles )
df
```

```python
df['Name'] = ''
```

```python
# Rearrange the columns with "Name" as the first column
new_order = ['Name'] + [col for col in df.columns if col != 'Name']
df = df[new_order]

# Display the DataFrame with the rearranged columns
print(df)
```

```python
df
```

```python
data_column = table.find_all('tr')
```

```python
for row in data_column:
    row_data= row.find_all('h6')
    individual_row_data = [data.text.strip() for data in row_data]
   
    length = len(df)
    df.loc[length] = individual_row_data
```
